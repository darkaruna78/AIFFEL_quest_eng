```python
class Square:
    def __init__(self):  # 생성자__init__의 매개변수의 첫번째는 self여야 합니다. 
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))  # 입력값을 int로 묶어야 합니다.

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')  # Square 클래스에 rect() 메서드를 추가합니다. 

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')  # Square 클래스에 par() 메서드를 추가합니다.  
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')  # Square 클래스에 trape() 메서드를 추가합니다. 
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    #직사가형 메서드 
    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    # 평행사변형 메서드
    def par(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '평행사변형의 넓이는 : ' + str(area)
        return result
    
    # 사다리꼴 메서드
    def trape(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result

            
```

```python

Kiosk 


class Kiosk: 
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.temp = ""
        self.order_price = []  # 이 리스트가 있어야 append가 가능합니다.
        self.order_menu = []
        self.price_sum = 0
                     
# 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')


 # 주문 메서드 1
    def menu_select(self):
        print()  # 줄 바꿈
        n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력
        self.price_sum = self.price[n-1]  # 선택 음료의 가격

        # 음료 온도 물어보기
        t = 0  # 기본값을 넣어주고
        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t==1 :
                temp = "HOT"
            elif t==2 :
                temp = "ICE"   
            else:    
                print("다시 입력하세요.\n")

        print('주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')  # 온도 속성을 추가한 주문 결과 출력

#추가 주문 
        while n != 0 :
            print()
            n = int(input("추가 주문은 음료 번호를 지불은 0을 누르세요 :"))
            if n >0 and n <len(self.menu) +1 :
                #추가음료 온도 
                t = 0  # 기본값을 넣어주고
                while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                    t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t==1 :
                        temp = "HOT"
                    elif t==2 :
                        temp = "ICE"   
                    else:    
                        print("1과 2 중 하나를 입력하세요.\n")
        
                self.order_price.append(self.price[n-1])  # 가격 리스트에 추가
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])  # 주문 리스트에 추가
                self.price_sum += self.price[n-1]
                
                print('주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원\n', '합계 :', self.price_sum,'원')  # 온도 속성을 추가한 주문 결과 출력
        else :
            if n == 0 : #지불을 입력하면
                print("주문이 완료되었습니다.")
                print(self.order_menu,self.order_price)
                pay(self)
            else :
                print("없는 메뉴입니다.")
    # 지불 메솓,  
    def pay(self):
        #합계 금액 출력
        print(f'총 합계 금액은 {self.price_sum}원 입니다.')
        #지불방법
        self.pay_method = 0 # 기본값
        while self.pay_method !=1 and self.pay_method !=2 : # 1,2 번을 입력하지 않으면 계속해서 반복
            self.pay_method = int(input('현금 결제는 1번, 카드 결제는 2번을 입력해주세요. :'))
            if self.pay_method == 1 :
                print('직원을 호출하겠습니다.')
            elif self.pay_method == 2 :
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else :
                print('다시 결제를 시도해 주세요.')

    def table(self):
        
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ':', self.order_price[i])
    
        print('합계 금액 :' , self.price_sum)
                
 # # 주문 메서드 2
 #    def menu_select(self):
 #        print()  # 줄 바꿈
 #        n = 0
 #        while n < 1 or len(menu) < n:
 #            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

 #            # 메뉴판에 있는 음료 번호일때
 #            if 1 <= n & n <= len(menu):
 #                self.price_sum = self.price[n-1]  # 선택 음료의 가격
        
 #            # 메뉴판에 없는 번호일 때
 #            else:  
 #                print("없는 메뉴입니다. 다시 주문해 주세요.")
```
